```python
import matplotlib.pyplot as plt
import numpy as np

# Data RMSE
models = ['SVD', 'KNN', 'Hybrid']
rmse = [0.1428, 0.1416, 0.267]

# Lebar bar
bar_width = 0.35
x = np.arange(len(models))

# Grafik RMSE
plt.figure(figsize=(8, 6))
plt.bar(x - bar_width / 2, rmse, width=bar_width, label='RMSE', color='lightgray', edgecolor='black', hatch='//')

# Label dan judul untuk RMSE
plt.xlabel('Model')
plt.ylabel('RMSE Value')
plt.title('Comparison of RMSE')
plt.xticks(x, models)
plt.legend()

# Tampilkan grafik RMSE
plt.tight_layout()
plt.show()
```
